```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df1 = pd.read_csv("charity_data.csv")
application_df.head(10)
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(['EIN', 'NAME'], axis=1).copy()

application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = []
app_counts = application_df['APPLICATION_TYPE'].value_counts()

for x in range(0,len(app_counts)):
    
    if app_counts[x] < 100:
        application_types_to_replace.append(app_counts.index[x])

# Replace in dataframe

for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts().loc[lambda x: x > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = []
class_counts = application_df['CLASSIFICATION'].value_counts()

for x in range(0,len(class_counts)):
    
    if class_counts[x] < 120:
        classifications_to_replace.append(class_counts.index[x])

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies_df = pd.get_dummies(application_df)

dummies_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = dummies_df.drop(['IS_SUCCESSFUL'], axis=1)
y = dummies_df['IS_SUCCESSFUL'].values



# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 49)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model 1

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=25, activation = 'relu', input_dim=len(X_train_scaled[0])))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=25, activation = 'relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation = 'sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Callback to save weights
from keras.callbacks import Callback
callback = tf.keras.callbacks.ModelCheckpoint(filepath="Checkpoints.{epoch}.hdf5", verbose=1, save_weights_only=True, save_freq="epoch", period=5, monitor='accuracy')
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=25, callbacks=[callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoup_Charity.hdf5")
```

## Model #2 - Bin the Ask Amount

```python
# Look at ASK_AMT value counts for binning
application_df['ASK_AMT'].value_counts()
```

```python
#  Bin the Ask Amounts to see if that would help the model.  I tried differnt binning amounts and it had little impact on the results

def replacing(x):
    if x <= 5000:
        return '<5001'
    elif x <= 50000:
        return '5001-50000'
    elif x <= 150000:
        return '50001-150000'
    elif x <= 500000:
        return '1500001-500000'
    elif x <= 1000000:
        return '500001-1000000'
    else:
        return '>1000000'

application_df['ASK_AMT']= application_df['ASK_AMT'].apply(replacing)
```
